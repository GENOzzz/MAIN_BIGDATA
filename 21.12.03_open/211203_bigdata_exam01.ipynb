# 판다스(Pandas) 사용법

```python
##pandas 기본 활용
import pandas as pd
names=['bob','jessica','mel','geno','gabdol','mary','john','cavin']
birth=[968,155,77,578,973,79,48,999]
custom=[1,5,25,13,23232,7777,9999,54576]
hoby=['run','movie','book','ani','game','sing','drum','piano']
                                              #list()
BabyDataSet=list(zip(names,birth,custom,hoby))#zip()
df=pd.DataFrame(data=BabyDataSet,columns=['Names','Births','Customs','Hobys'])#DataFrame 객체.
df.head()#데이터 프레임의 상단 부분 출력
```

#### list() : 파이썬 함수 d=list() 빈배열 생성.
#### zip() : 리스트, 튜플 같이 반복 가능한 자료형 개수가 동일할 때 사용한다.  
자료형의 각각의 요소를 나눈 후 순서대로 묶어서 요소 개수만큼 새로운 iterable 자료형을 생성
#### DataFrame : Row, Column, Series 들로 구성

# 넘파이(Numpy) 사용법

##### Numerical Python의 줄임말로 수치 계산을 위해 만들어진 라이브러리

```python
#배열 생성 , 이 배열은 넘파이 배열이며 파이썬의 기본 자료구조와는 다른 데이터타입이다.
import numpy as np
arr1=np.arange(15).reshape(3,5)
arr1
```

```python
arr2_0=np.arange(1,100,2)#한줄배열
arr2=np.arange(1,100,2).reshape(5,10)#5x10배열
arr2_1=np.arange(1,100,2).reshape(10,5)#10x5배열
print('arr2_0=')
print(arr2_0)
print('arr2=')
print(arr2)
print('arr2_1=')
print(arr2_1)
```

### 넘파이 배열 정보 확인하기

```python
arr1.dtype
```

### 다른 형태의 배열 생성하기

```python
arr3=np.zeros((3,4))
print('arr3=')
print(arr3)
arr3_1=np.zeros((3,4,5))
print('arr3_1=')
print(arr3_1)
arr3_2=np.ones((3,4,5))
arr3_2[1][3][2]=7
print('arr3_2=')
print(arr3_2)
```

### 넘파이 배열을 이용한 사칙연산 수행하기

```python
arr4 =np.array([
    [1,2,3],
    [4,5,6]
],dtype=np.float64)
arr5 =np.array([
    [7,8,9],
    [10,11,12]
],dtype=np.float64)
#사칙연산을 출력합니다.
print("arr4+arr5=")#넘파이 배열끼리의 더하기
print(arr4+arr5,"\n")#각 자리끼리 더해서 나옴 ==> 동일한 shape이여야 함
print("arr4-arr5=")#넘파이 배열끼리의 빼기
print(arr4-arr5,"\n")
print("arr4*arr5=")#넘파이 배열끼리의 곱하기
print(arr4*arr5,"\n")
print("arr4/arr5=")#넘파이 배열끼리의 나누기
print(arr4/arr5,"\n")
print(arr4[0][0])#해당 원소 출력
```

# Matplotlib 사용하기

```python
%matplotlib inline
import matplotlib.pyplot as plt

y=df['Births']
x=df['Names']

#막대그래프
plt.bar(x,y)# 막대 그래프 객체 생성
plt.xlabel('Names')#X축 제목
plt.ylabel('Births')#Y축 제목
plt.title('Bar plot')#그래프제목
plt.show()#그래프 출력
```

```python
#산점도그래프
#랜덤 추출 시드를 고정
np.random.seed(19920613)

#산점도 데이터를 생성
x=np.arange(0.0, 100.0, 5.0)
y=(x*1.5)+np.random.rand(20)*50
x1=np.arange(0.0, 100.0, 2.0)
y1=(x1*1.5)+np.random.rand(50)*50
#산점도데이터출력
plt.scatter(x, y, s=y,c="b",alpha=0.5, label="scatter point")
plt.scatter(x1,y1, s=x1,c="r", alpha=1, label="scatter point1")
plt.xlabel("x")#x축이름
plt.ylabel("y")#y축이름
plt.legend(loc='upper left')#범례표시(위치)
plt.title("Scatter plot")#그래프이름
plt.show()
```

### scatter
x,y=좌표/
s=size(숫자로)/
c=color/
alpha=transparency (투명도)/
cmap=colormap (컬러맵에 해당하는 문자열지정)/
