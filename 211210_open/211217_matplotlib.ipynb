### Matplotlib기본 사용

matplotlib.pyplot 모듈은 각각의 함수를 사용해 그래프를 만들고 변화를 줄 수 있다.

### 기본 그래프 그리기



#### 예제 1

```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4])
plt.show()
```

pyplot.plot()함수에 하나의 숫자 리스트르 입력함으로써 위와 같은 그래프가 그려짐.

plot() : 리스트의 값들이 y좌표의 값 들이라고 가정하고, x 좌표의 값[0,1,2,3]을 자동으로 만들어냄.

show() : 그래프를 화면에 나타나도록 함.

```python
plt.plot([1,2,3,4],[1,4,9,16])
plt.show()
```

plot() 함수는 임의의 개수의 인자를 받을수 있다. (X-Y값을 차례로)

```python
plt.plot([1,2,3,4],[1,4,9,16],'ro')
plt.axis([0,6,0,20])
plt.show()
```

x,y인자 값에 대해 선의 색상과 형태를 지정하는 포맷문자열(Format string)을 세번째 인자에 입력.

'ro' : red,원형(o) 마커를 의미

'b-' : blue,실선(-) 을 의미

axis([xmin,xmax,ymin,ymax])지정 가능

```python
import numpy as np
#200ms 간격으로 균일하게 샘플된 시간
t=np.arange(0.,5.,0.2)
#빨간 대쉬,파란 사각형,녹색 삼각형
plt.plot(t,t,'r--',t,t**2.5,'bs',t,t**3,'g^')
plt.show()
```

Numpy어레이를 이용한 방법

Numpy어레이를 사용하지 않더라도 모든 시퀀스는 내부적으로 Numpy어레이로 변환.

이 예제는 다양한 스타일을 갖는 여러 개의 곡선 하나의 그래프로 나타냄.
