[검색어에 따라 다른 결과 나타내기]

+ 프로토콜,도메인,패스(경로),파라미터(서버의 추가적인 정보를 제공하기 위해서 사용, 키와 값으로 구성,&으로 구분)

+ query는 검색어에 해당됨

+ 파이썬에서 사용자에게 검색어를 입력받은 뒤 query값에 적용하면 검색어 변경가능

```python
keyword=input("insert keyword : ")
respon=requests.get("https://search.naver.com/search.naver?where=news&sm=tab_jum&query="+keyword)
html=respon.text
soup=bsp(html,'html.parser')
links=soup.select('.news_tit')
for link in links:
    tit=link.text
    scr=link.attrs['href']
    print(tit,scr)
```

### [실전 프로젝트 : 뉴스데이터 수집하기 여러페이지 결과 모두 가져오기]
- URL 분석하기, 여러 페이지의 결과를 가져오기 위해 이동하면서 URL이 어떻게 변하는지 살펴보기
- 뉴스 다음페이지로 이동시 start 파라미터의 값이 10씩 증가
- 1페이지 : start =1, 2페이지 : start=11, 3페이지 : start=21...n페이지 : strat = nx10+1

```python
intCheck=False
keyword=pag.prompt("검색어 입력")
while(not intCheck):
    lastPage=pag.prompt("마지막페이지 입력")
    intCheck=lastPage.isnumeric()#isnumberic() : 입력받은 숫자가 양의 정수인지 유효성 검사.
page=0
num=0
lp=int(lastPage)
for i in range(1,(lp*10),10,):
    page+=1
    respon=requests.get(f"https://search.naver.com/search.naver?where=news&sm=tab_jum&query={keyword}&start={i}")
    html=respon.text
    soup=bsp(html,'html.parser')
    links=soup.select('.news_tit')
    print(f"============================{page}페이지============================")
    for link in links:
        num+=1
        tit=link.attrs['title']
        src=link.attrs['href']
        print(f"{num}. {tit}",src)
```
