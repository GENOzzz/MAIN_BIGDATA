```python
import requests
from bs4 import BeautifulSoup

url = 'http://www.yjjob.or.kr/p/?j=23'

res = requests.get(url)
soup = BeautifulSoup(res.content, 'html.parser')

print(soup.prettify())
```

```python
tds = soup.find_all('td', 'lh_140 v_top relative')
```

```python
for td in tds:
    # 과정 구분과 과정명을 가져오기 위한 a 태그 선택
    td_a = td.find('a', 'p100')
    # 과정구분
    td_a_em = td_a.find('em', 'font11 gray')
    print('과정구분 :', td_a_em.text.strip())
    # 과정명
    td_a_span = td_a.find('span', 'font14 bold')
    print('과정명 :', td_a_span.text.strip())    
    
```

```python
for td in tds:
    # 교육기간
    edu_s = td.find('em', 'blue')
    edu_e = td.find('em', 'red')
    # print(edu_s.text.strip())
    # print(edu_e.text.strip())
    ### 만약 교육기간이 없을 경우 (edu_s, edu_e 가 null 인 경우) 오류 발생
    print('교육기간 :', edu_s.text.strip(), '~', edu_e.text.strip()) if edu_s else print('교육기간 :', '') 
```

```python
for td in tds:
    # 그 외
    # 모두 p 태그에 class가 tpad5의 속성값을 가진다
    td_ps = td.find_all('p', 'tpad5')
    
    for td_p in td_ps:
        td_p_text = str(td_p.text)
        # text 형식 => 훈련코드 : xxxxx
        print(td_p_text)
    print('-------')
```

```python
for td in tds:
    td_ps = td.find_all('p', 'tpad5')
    
    for td_p in td_ps:
        td_p_text = str(td_p.text)
        # text 형식 => 훈련코드 : xxxxx
        # ':'를 기준으로 앞의 key와 뒤의 value를 분리
        key_str = td_p_text[:td_p_text.find(':')].strip()
        val_str = td_p_text[td_p_text.find(':')+1:].strip()
        print(key_str, ':', val_str)  
```

```python
for td in tds:
    td_ps = td.find_all('p', 'tpad5')
    
    for td_p in td_ps:
        td_p_text = str(td_p.text)
        # value의 tab을 replace
        key_str = td_p_text[:td_p_text.find(':')].strip()
        val_str = td_p_text[td_p_text.find(':')+1:].replace('\t', '').strip()
        print(key_str, ':', val_str)    
```

```python
for td in tds:
    # 훈련코드
    tran_cd = td.find('span', 'red bold')
    print('훈련코드 :', tran_cd.text.strip())
```

```python
# 위의 오류에서 훈련코드가 없는 과정일 경우 null 객체에 text속성과 strip()를 적용하려니 오류가 발생한다.
# null 처리를 해 주어야 한다.
for td in tds:
    # 훈련코드
    tran_cd = td.find('span', 'red bold')
    print('훈련코드 :', tran_cd.text.strip()) if tran_cd else print('훈련코드 :', '')
```

```python
# 과정구분, 과정명 , 그 외 -> 다른 방법
tds = soup.find_all('td', class_='lh_140 v_top relative')

for td in tds:
    td_a = td.find('a', 'p100')
    # 과정구분
    td_a_em = td_a.find('em', 'font11 gray')
    print('과정구분 :', td_a_em.text.strip())
    # 과정명
    td_a_span = td_a.find('span', 'font14 bold')
    print('과정명 :', td_a_span.text.strip())
    # 교육기간
    edu_s = td.find('em', 'blue')
    edu_e = td.find('em', 'red')
    print('교육기간 :', edu_s.text.strip(), '~', edu_e.text.strip()) if edu_s else print('교육기간 :', '')     
    
    # 그 외
    td_ps = td.find_all('p', 'tpad5')
    
    for td_p in td_ps:
        td_p_text = str(td_p.text)
        key_str = td_p_text[:td_p_text.find(':')].strip()
        val_str = td_p_text[td_p_text.find(':')+1:].replace('\t', '').strip()
        print(key_str, ':', val_str)
```

```python
list_all = []
dic_row = {}

# 과정구분, 과정명 , 그 외 -> 다른 방법
tds = soup.find_all('td', class_='lh_140 v_top relative')

for td in tds:
    td_a = td.find('a', 'p100')
    # 과정구분
    td_a_em = td_a.find('em', 'font11 gray')
    # print('과정구분 :', td_a_em.text)
    dic_row['과정구분'] = td_a_em.text.strip()
    # 과정명
    td_a_span = td_a.find('span', 'font14 bold')
    # print('과정명 :', td_a_span.text)
    dic_row['과정명'] = td_a_span.text.strip()
    # 교육기간
    edu_s = td.find('em', 'blue')
    edu_e = td.find('em', 'red')
    # print('교육기간 :', edu_s.text.strip(), '~', edu_e.text.strip()) if edu_s else print('교육기간 :', '')     
    if edu_s:
        dic_row['교육기간'] = edu_s.text.strip(), '~', edu_e.text.strip()
    
    # 그 외
    td_ps = td.find_all('p', 'tpad5')
    
    for td_p in td_ps:
        td_p_text = str(td_p.text)
        key_str = td_p_text[:td_p_text.find(':')].strip()
        val_str = td_p_text[td_p_text.find(':')+1:].replace('\t', '').strip()
        # print(key_str, ':', val_str)
        dic_row[key_str] = val_str
    
    list_all.append(dic_row)     
```
